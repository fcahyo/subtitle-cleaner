#### 1. Import library regex

```python
import re
```

#### 2. Membaca file subtitle

```python
with open('subtitle.txt', 'r') as file:    
    text = file.read()
```

#### 3. Mengubah file yang awalnya bertipe list menjadi string agar regex bisa digunakan

```python
subtitle = ''.join(text)
```

```python
subtitle
```

#### 4. Regex untuk menghilangkan nomor baris dan penanda waktu
- '\d+' berarti nomor dengan digit satu atau lebih diikuti dengan '\n' yang artinya baris baru
- Untuk penanda waktu dikelompokkan menjadi 3 grup (Grup dalam regex ditulis di dalam tanda kurung '()')
- Grup pertama dan ketiga berisi waktu mulai dan waktu berakhir yang ditulis dengan (\d{2}:\d{2}:\d{2},\d{3})
- Tanda '{}' mengartikan banyak digit dalam nomor
- Untuk grup kedua berisi .* yang berarti string apapun, dalam kasus ini yaitu tanda panah

```python
subtitle = re.sub(r'\d+\n(\d{2}:\d{2}:\d{2},\d{3})(.*)(\d{2}:\d{2}:\d{2},\d{3})', '', subtitle)
```

#### 5. Regex untuk menghilangkan iklan
- Iklan dalam subtitle ini ditandai dengan adanya tag font maka dibuat regex untuk menghilangkan tag font beserta isinya

```python
subtitle = re.sub(r'<font.*', '', subtitle)
```

#### 6. Regex untuk menghilangkan tag html
- Tag html diawali dengan '<' dan diakhiri dengan '>'
- '[^>]' berarti hanya mencocokkan yang tidak memiliki karakter '>'
- Diikuti dengan * agar tag selanjutnya juga dicek

```python
subtitle = re.sub(r'(<[^>]*>)', '', subtitle)
```

#### 7. Hasil dari cleaning

```python
subtitle
```

#### 8. Menghapus newline yang tidak perlu
Dikarenakan <b>re.sub()</b> dari python mengubah kalimat yang ingin dihapus menjadi newline, maka dibutuhkan regex lagi untuk menghapus newline yang terlalu banyak dan tidak perlu.

```python
subtitle = re.sub('(\\n\\n\\n\\n\\n)', '', subtitle)
subtitle = re.sub('(\\n\\n)', '', subtitle)
```

#### 9. Menulis hasil cleaning ke dalam file d_subtitle.txt

```python
with open('d_subtitle.txt', 'w') as f:
    f.write(subtitle)
```
